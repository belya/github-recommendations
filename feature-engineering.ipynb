Предварительно запустить dataset/extract_small.sh

```python
import pandas as pd
github = pd.DataFrame()
```

```python
import json

created = []
types = []
actors = []
repos = []
messages = []
jsons = []

with open("./dataset/2017-12-06-14.json") as file:
    for event in file:
        event = json.loads(event)
        created.append(event["created_at"])
        types.append(event["type"])
        actors.append(event["actor"]["login"])
        repos.append(event["repo"]["name"])
        jsons.append(event)
```

```python
github["created"] = created
github["type"] = types
github["actor"] = actors
github["repo"] = repos
github["json"] = jsons
```

```python
for type in github["type"].unique():
    print(type)
    print(github[github["type"] == type]['json'].iloc[0])
    print("=====================================")
```
